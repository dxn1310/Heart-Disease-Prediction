```python
#importing the required modules
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('heart_data.csv') #reading the data
```

Data Preprocessing

```python
print(df.head())
```

```python
df = df.drop('Unnamed: 0',axis=1)
```

```python
print(df.head())
```

```python
#Plotting the graphs 
sns.lmplot(x='biking',y='heart.disease',data=df)
```

```python
sns.lmplot(x='smoking',y='heart.disease',data=df)
```

Creating Training data

```python
x_df =df.drop("heart.disease",axis=1)
y_df = df['heart.disease']
```

```python
print(x_df.head()),print(y_df.head())
```

Model Training

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.3, random_state=42) #splitting the data into train and test data
```

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
print(model.score(X_train,y_train))
```

```python
prediction_test = model.predict(X_test)  
```

```python
print(y_test, prediction_test)
```

```python
prediction_test[0]
```

```python
y_test[0]
```

```python
print("Mean sq. errror between y_test and predicted =", np.mean(prediction_test-y_test)**2)
```

Saving the model using pickle

```python
import pickle
pickle.dump(model, open('models/model.pkl','wb'))

model = pickle.load(open('models/model.pkl','rb'))
print(model.predict([[20.1, 56.3]]))
```
